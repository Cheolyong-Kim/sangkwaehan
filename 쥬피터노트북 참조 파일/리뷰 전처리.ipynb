```python
import pandas as pd
import numpy as np
import pickle
import os
```

```python
file_list = os.listdir(os.getcwd()+'\\merged_pickle')
```

```python
for file_name in file_list:
    with open(f'merged_pickle/{file_name}', 'rb') as fr:
        data = pickle.load(fr)
    df = pd.DataFrame(data['총리뷰'], columns=['리뷰', '평점'])
    df.dropna(subset=['리뷰'], how='any', axis=0, inplace=True)
    median_star = df['평점'].median()
    df['평점'] = df['평점'].fillna(median_star)
    
    df['리뷰'] = df['리뷰'].str.replace('[^ㄱ-ㅎ ㅏ-ㅣ 가-힣]', '')
    df['리뷰'] = df['리뷰'].str.replace('^ +', '')
    df['리뷰'] = df['리뷰'].replace('', np.nan)
    
    df.dropna(subset=['리뷰'], how='any', axis=0, inplace=True)
    
    data['총리뷰'] = df.values.tolist()
    
    with open(f'p_merged_pickle/{file_name}', 'wb') as fw:
        pickle.dump(data, fw)
```
